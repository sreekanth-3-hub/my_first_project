```python
#Iterating numpy arrays
import numpy as np
arr1 = np.array(['a','b','c'])
ls2 = [[1,2,3,4], [4,5,6, 7], [7,8,9,3]]
arr2 = np.array(ls2) #2D array = Matrix
print(arr1)
print(arr2)
```

```python
for row in range(3):
  for col in range(4):
    print(arr2[row, col], end=' ') #Indexing
  print("\n")
```
